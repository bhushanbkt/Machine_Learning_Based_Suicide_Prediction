```python
import numpy as np

# Define suitable conditions to categorize suicide risk
low_risk_condition = (df['suicides/100k pop'] <= 5.0)
medium_risk_condition = ((df['suicides/100k pop'] > 5.0) & (df['suicides/100k pop'] <= 15.0))
high_risk_condition = (df['suicides/100k pop'] > 15.0)

# Create a new variable 'suicide_risk' based on conditions
df['suicide_risk'] = np.where(low_risk_condition, 'Low', np.where(medium_risk_condition, 'Medium', 'High'))
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['suicide_risk'] = label_encoder.fit_transform(df['suicide_risk'])
```
